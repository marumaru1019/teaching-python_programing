### 実際にそれぞれの型について見ていきましょう。

````python
#数値型について
#数値型にはint型とfloat型という２つの型があります。
#intはinteger(整数)、floatはfloating-point number(浮動小数点)からきています。

num1=1
num2=0.3
````

````python
#型はtype(変数)で調べることが出来ます。
print(num1)
print(type(num1))
print(num2)
print(type(num2))
````

````python
#文字型について
#文字型はstr型がありstringからきています。
#変数を宣言するときは""をつけて宣言します。

string = "文字ですよ"
print(string)
print(type(string))
````

````python
#真偽型について
#真偽型はbool型がありBooleanからきています。(booleanは<はい:いいえ><ON:OFF>みたいなもの)

true = 1<2
false = 1>2
print(true)
print(type(true))
print(false)
print(type(false))
````
